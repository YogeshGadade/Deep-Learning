# Reading and printing o/p of tesseract for the images used in this example:
- one hindi language sample
- one english language sample

```python
# this line of code is used to read the image on pytesseract site itself. lets try it
try:
    from PIL import Image   
except ImportError:
    import Image      
import pytesseract
```

```python
image = Image.open('English sample 2.jpg')
image   
```

```python
EnglishText = pytesseract.image_to_string(image, lang='eng')
print(EnglishText)
```

```python
image = Image.open('hindi sample with different fonts.png')
image   
```

```python
HindiText = pytesseract.image_to_string(image, lang='hin')
print(HindiText)
```
