# Goal: Using Google Vision API word detection i.e. OCR

- Input: English and Hindi Text images
- Output: language recognition, Text recognision 

Requirements: enabled Google vision API (refer to this article know how to do it: https://limitlessdatascience.wordpress.com/2019/02/10/installation-of-google-cloud-vision-api/)

For more detailed explaination please refer this article:  https://limitlessdatascience.wordpress.com/2019/09/19/vision-api-detect-handwriting-ocr-python-code-implementation/

```python
# import packages
import PIL
print("PIL version:", PIL.__version__)
import PIL.Image as Image
from google.cloud import vision
import io
```

```python
def detect_document(path):
    """Detects document features in an image."""
    client = vision.ImageAnnotatorClient()

    with io.open(path, 'rb') as image_file:
        content = image_file.read()

    image = vision.types.Image(content=content)
    response = client.document_text_detection(image=image) #response is in JSON format
    
    #print the response to the input image by vision API: block confidence
    for page in response.full_text_annotation.pages:
        for block in page.blocks:
            print('\nBlock confidence: {}\n'.format(block.confidence))

            for paragraph in block.paragraphs:
                print('Paragraph confidence: {}'.format(
                    paragraph.confidence))

                for word in paragraph.words:
                    word_text = ''.join([
                        symbol.text for symbol in word.symbols
                    ])
                    print('Word text: {} (confidence: {})'.format(
                        word_text, word.confidence))

                    for symbol in word.symbols:
                        print('\tSymbol: {} (confidence: {})'.format(
                            symbol.text, symbol.confidence))
```
